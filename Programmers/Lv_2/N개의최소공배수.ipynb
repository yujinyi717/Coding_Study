1. 최대공약수

- 정의 : A,B 값이 주어졌을 때, 공통이 되는 약수 중에서 최대값을 의미
- 구하는 법 :
    - A,B 각각의 약수를 구해서 공통이되는 약수 중에서 가장 큰 값을 찾는 방법 -> 찾지 않아도 되는 약수들까지 구해야함 (비효율적)
    - 유클리드 호제법 (연산이 빨라 효율적)
         - gcd(A,B) = gcd(B,A%B)
         - 아래 예시처럼 점점 줄어들게 되고, 0 등장시 왼쪽 값 출력 ( 0의 약수는 모든 수 이기 때문 ) 
      
2. 최소공배수

- A x B = gcd*2 x a x b = 최소공배수 x 최대공약수
- 따라서, 최소공배수 = A x B / 최대공약수

### 참고, 두개의 인자일경우

```python
# 패키지 사용 x 

def gcd(a,b): # 최대공배수
    while b > 0 :
        a,b = b, a%b
    return a

def lcm(a,b):
    return a*b//gcd(a,b)
lcm(2,4) # 최소공배수
```

3. 문제풀이

```python
# 모듈 사용 x 

# 최대공약수 
def gcd(a,b):
    while b > 0:
        a,b = (b, a%b)
    return a

# 최소공배수
def lcm(arr):
    answer = arr[0]
    for i in arr:
        answer = answer*i//gcd(answer,i)
    return answer
print(lcm([2,6,8,14]));  
```

```python
# 모듈 사용 o 

# 최소공배수
import math

def lcm(arr):
    result = arr[0]
    for n in arr:
        result = result*n//math.gcd(n,result)
    return result
print(lcm([2,6,8,14]));  
```
